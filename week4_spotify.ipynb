```python
import pandas as pd 
import pandas_profiling
```

```python
df = pd.read_csv('data.csv', index_col=0)
df.head(2)
```

```python
df.describe()
```

```python
profile = df.profile_report(title='Spotify Dataset Analysis')
profile.to_file(output_file="Spotify Dataset Analysis.html")
```

```python
year_df = df.groupby(['year']).mean()
year_df = year_df.reset_index()
year_df.head(5)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

for column_name in list(year_df.columns.values):
    fig.add_trace(go.Scatter(
        x = year_df['year'],
        y = year_df[column_name],
        name=column_name,
        visible='legendonly'
    ))

fig.update_layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=True,
        showticklabels=True,
    ),
    autosize=False,
    showlegend=True,
    plot_bgcolor='white'
)
annotations = []

annotations.append(dict(xref='paper', yref='paper', x=0.0, y=1.05,
                              xanchor='left', yanchor='bottom',
                              text='Mean Characteristics Over Time',
                              font=dict(family='Arial',
                                        size=20,
                                        color='rgb(37,37,37)'),
                              showarrow=False))

fig.update_layout(annotations=annotations)

fig.show()
```
